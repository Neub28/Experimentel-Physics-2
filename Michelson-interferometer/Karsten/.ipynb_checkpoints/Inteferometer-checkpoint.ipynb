```python
entries_list = os.listdir("C:/Users/Karsten/Eksperimental/Experimentel-Physics-2/Michelson-interferometer/Karsten/Datafiler")


def get_ts_Vs(entries):
    ts = []
    Vas = []
    Vbs = []
    for entry in entries:
        data = io.loadmat(f"C:/Users/Karsten/Eksperimental/Experimentel-Physics-2/Michelson-interferometer/Karsten/Datafiler/{entry}")
        Vb = np.transpose(data.get("B"))[0]
        Va = np.transpose(data.get("A"))[0]
        t = np.linspace(0,100, len(Vb))
        Vas.append(Va)
        Vbs.append(Vb)
        ts.append(t)
    return ts, Vas, Vbs

def fit_val(x, a, b):
    f = a*x + b
    return f

```

```python
ts1, Vas1, Vbs1 = get_ts_Vs(entries_list)
print(len(Vas1))
```
