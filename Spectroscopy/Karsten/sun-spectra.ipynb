```python
rel_path = "../Spectrum of sun"
entries_list = os.listdir(rel_path)
print(entries_list)
```

```python
def get_data(rel_path, atom=""):
    data = np.loadtxt(f"{rel_path}/{atom}.txt",skiprows=1)
    Is = np.zeros(len(data))
    vws = np.zeros(len(data))
    for i, dat in enumerate(data):
        Is[i] = dat[1]
        vws[i] = dat[0]
    return vws, Is
```

```python
vw0, I0 = get_data(rel_path, "Sun1")
vw1, I1 = get_data(rel_path, "pære1")
vw2, I2 = get_data(rel_path, "pære2")
vw3, I3 = get_data(rel_path, "pære3")

def plaw(wl, *p) : 
    c = 299792458 # m/s
    h = 6.62607015e-34 # J*s
    kB = 1.380649e-23 # J/K
    A = p[0]
    T = p[1] # K
    return A*(2*h*c**2)/(wl**5)*1/(np.exp((h*c)/(wl*kB*T))-1)
```
