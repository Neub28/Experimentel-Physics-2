```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta_1 = np.linspace(0.001, np.pi/2, 5000)
n_air = 1
n_glass = 1.5

theta_ag = np.arcsin(n_air/n_glass*np.sin(theta_1))
theta_ga = np.arcsin(n_glass/n_air*np.sin(theta_1))
```

```python
def R(theta_1, theta_2):
    R_s = np.sin(theta_1 - theta_2)**2/(np.sin(theta_1 + theta_2)**2)
    R_p = np.tan(theta_1 - theta_2)**2/(np.tan(theta_1 + theta_2)**2)
    return R_s, R_p
```

```python
R_ag = R(theta_1, theta_ag)
R_ga = R(theta_1, theta_ga)
Rs = [R_ag, R_ga]
```

```python
fig, axs = plt.subplots(1,2, figsize=(12,6))

for ax, Rl in zip(axs, Rs):
    ax.tick_params(direction="out", labelsize=15)
    ax.grid()
    ax.set_xlabel(r"$\theta_1$", fontsize=18)
    ax.set_ylabel("$R$", fontsize=18)
    for R in Rl:
        ax.plot(theta_1, R, lw=2)
axs[0].set_title("For luft til glas", fontsize=24)
axs[1].set_title("for glas til luft", fontsize=24)
```
