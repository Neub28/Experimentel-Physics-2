```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import scipy.stats as stats
import scipy.io as io
import time
from os import listdir
from scipy.signal import find_peaks

from scipy.interpolate import CubicSpline

plt.rc("font", family=["Times New Roman"]) #skifter skrifttype
plt.rc("axes", labelsize=18)   #skriftstørrelse af `xlabel` og `ylabel`
plt.rc("xtick", labelsize=16, direction="out")  #skriftstørrelse af ticks og viser ticks øverst
plt.rc("ytick", labelsize=16, direction="out")
plt.rc("axes", titlesize=22)
plt.rc("legend", fontsize=16)
```

```python
def getdata(string) : 
    rpath = r"../Spectral lamps/"
    data = np.genfromtxt(rpath+string, skip_header=1, delimiter='\t')
    l = []
    i = []
    for ar in data : 
        l.append(ar[0])
        i.append(ar[1])
    return (l,i)
    
```

```python
slist = listdir(r"../Spectral lamps/")
print(slist)

def writer(ll, il, pl, atom) : 
    file = open(f"{atom}-data.txt", 'w')
    for i in range(len(ll)) : 
        if i > len(pl)-1 : 
            file.write("{} \t {} \n".format(ll[i], il[i]))
        else : 
            index = pl[i]
            file.write("{} \t {} \t {} \t {} \n".format(ll[i], il[i], ll[index], il[index]))
        
    file.close()
            
```

# Argon

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.grid()

(lar, iar) = getdata(slist[0])
ax.plot(lar, iar)

peaks, _ = find_peaks(iar, height=5000)

lar_peaks = []

for p in peaks : 
    lar_peaks.append(lar[p])
    ax.plot(lar[p], iar[p], '*', markersize=10)
    
print(lar_peaks)

writer(lar, iar, peaks, "Ar")

```
